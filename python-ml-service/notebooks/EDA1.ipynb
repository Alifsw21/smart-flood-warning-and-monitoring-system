```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/Flood_Prediction_NCR_Philippines.csv")

kolom = ['Date', 'Location', 'Rainfall_mm', 'WaterLevel_m', 'SoilMoisture_pct', 'FloodOccurrence']
df_filtered = df[kolom].copy()

df_filtered['Date'] = pd.to_datetime(df_filtered['Date'])
df_urut = df_filtered.sort_values('Date')

plt.figure(figsize=(14, 5))

plt.plot(df_urut['Date'], df_urut['WaterLevel_m'], color='dodgerblue', linewidth=1.5, label='Ketinggian Air (m)')

plt.title('Tren Pergerakan Ketinggian Air Sungai di Filipina', fontsize=13, pad=12)
plt.xlabel('Tanggal Analisis', fontsize=11)
plt.ylabel('Ketinggian Air (Meter)', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()

plt.figure(figsize=(14, 5))

plt.plot(df_urut['Date'], df_urut['SoilMoisture_pct'], color='saddlebrown', linewidth=1.5, label='Kelembapan Tanah (%)')

plt.title('Tren Pergerakan Kelembapan Tanah (Soil Moisture) di Filipina', fontsize=13, pad=12)
plt.xlabel('Tanggal Analisis', fontsize=11)
plt.ylabel('Rata-rata Kelembapan Tanah (%)', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```

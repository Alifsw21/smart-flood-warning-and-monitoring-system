```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/data_finish.csv")

kolom = ['date', 'Tn', 'Tx', 'Tavg', 'RH_avg', 'RR', 'ss', 'ff_x', 'ddd_x', 'ff_avg', 'ddd_car', 'region_name', 'flood']
df_filtered = df[kolom].copy() 

df_filtered['date'] = pd.to_datetime(df_filtered['date'])
df_urut = df_filtered.sort_values('date')

plt.figure(figsize=(14, 5))

plt.plot(df_urut['date'], df_urut['RH_avg'], color='teal', linewidth=1.5, label='Kelembapan Udara (%)')

plt.title('Tren Pergerakan Kelembapan Udara (RH_avg) di Indonesia', fontsize=13, pad=12)
plt.xlabel('Tanggal Analisis', fontsize=11)
plt.ylabel('Rata-rata Kelembapan (%)', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()

plt.figure(figsize=(14, 5))

plt.plot(df_urut['date'], df_urut['RR'], color='crimson', linewidth=1.5, label='Curah Hujan (mm)')

plt.title('Tren Pergerakan Intensitas Curah Hujan (RR) di Indonesia', fontsize=13, pad=12)
plt.xlabel('Tanggal Analisis', fontsize=11)
plt.ylabel('Volume Curah Hujan (mm)', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```
